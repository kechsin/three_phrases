```python
import torch
from torch.utils.data import DataLoader

import pandas as pd

import pickle
data = pd.read_pickle("balanced_vectorized.pkl")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = (stopwords.words('english'))
nltk.download('punkt')
```

```python
def get_unique_chars(x, column):
  unique_chars = []
  for index, row in x.iterrows():
    if str(row[column]).lower() not in unique_chars:
      unique_chars.append(str(row[column]).lower())
  return(unique_chars)
```

```python
from itertools import chain
```

```python
get_unique_chars(data, 'character')
```

```python
all = []
for i in get_unique_chars(data, 'character'):
  j = word_tokenize(i)
  all.append(j)
print(all)
```

```python
flat_list = list(chain.from_iterable(all))
```

```python
flat_list.append('lord')
flat_list.append('nagini')
```

```python
stop_words_eng = stop_words + flat_list
```

```python
for index, row in data.iterrows():
  if len(str(row['text'])) < 4:
      stop_words_eng.append(str(row['text']))
```

```python
data['goodness'] = data['goodness'].apply(lambda x: x.replace('Trash', 'Neutral'))
data['text'] = data['text'].apply(lambda x: ' '.join(word.lower() for word in word_tokenize(x) if word.isalnum()))
```

```python
train, test = train_test_split(data, test_size=0.1, shuffle=True)
train.reset_index(inplace=True)
test.reset_index(inplace=True)
```

```python
#vectorizer = CountVectorizer()
# X = np.vstack(train.vector.values)
# X_test = np.vstack(test.vector.values)
# y = train.goodness.values
# y_test = test.goodness.values
```

```python
vectorizer_stop = CountVectorizer(max_features=10000, min_df=5, max_df=0.4, ngram_range=(2,10), stop_words=stop_words_eng, tokenizer=word_tokenize)
X = vectorizer_stop.fit_transform(train.text)
X_test = vectorizer_stop.transform(test.text)
y = train.goodness.values
y_test = test.goodness.values
```

```python
X_test
```

```python
clf_Logistic = RandomForestClassifier(n_estimators=10000, criterion='log_loss', max_depth=10000, min_samples_split=10 )
clf_Logistic.fit(X, y)
preds_Logistic = clf_Logistic.predict(X_test)
print(classification_report(y_test, preds_Logistic, zero_division=0))

feature_Logistic = vectorizer_stop.get_feature_names_out()
feature_Logistic_importance = clf_Logistic.feature_importances_
kind = [feature_Logistic[i] for i in feature_Logistic_importance.argsort()[-5:][::-1]]
print(kind)
```
